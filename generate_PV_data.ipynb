# **Generate Thick walled pressure vessel data**


Vary inner and outer pressures, and inner and outer radii of the PV

```
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
pressure_inside = np.arange(5e6, 12e6, 1e6)
pressure_outside = np.arange(0.97e5, 1.04e5, 0.01e5)
thickness = 8*0.0254
radius_outer = np.arange(40, 44.5, 0.5)*0.0254
radius_inner = radius_outer - thickness
```

```
num_obs = len(pressure_inside)*len(pressure_outside)*len(radius_inner)*len(radius_outer)
PV_data = np.zeros((6, num_obs))
```

```
m = 0
for ro in radius_outer:
  for ri in radius_inner:
    for po in pressure_outside:
      for pi in pressure_inside:
        PV_data[0, m] = ro
        PV_data[1, m] = ri
        PV_data[2, m] = po
        PV_data[3, m] = pi
        m = m + 1
```

```
ro = PV_data[0, :]
ri = PV_data[1, :]
po = PV_data[2, :]
pi = PV_data[3, :]

A = (pi*(ri**2) - po*(ro**2))/(ro**2 - ri**2)
B = ((ro**2)*(ri**2)*(po - pi))/(ro**2 - ri**2)

sigma_c = A - (B)/((0.5*ri + 0.5*ro)**2)
sigma_r = A + (B)/((0.5*ri + 0.5*ro)**2)

PV_data[4, :] = sigma_c
PV_data[5, :] = sigma_r
```

```
np.savetxt("PV_data.csv", PV_data, delimiter=",")
```
